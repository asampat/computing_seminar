# Brewers Problem

Ref: The Allocation of Resources by Linear Programming, Scientific American, by Bob Bland.

A Brewery Produces Ale and Beer:
- Production limited by scarce resources: corn, hops, barley, malt



Goal: Find the number of ales and beer that should be brewed to maximize profit. 

```julia
using JuMP, Cbc
```

```julia
m = Model(solver = CbcSolver())

@variable(m, A >=0)
@variable(m, B >=0)
@variable(m, profit)

@constraint(m, corn_constraint, 5A + 15B <= 480)
@constraint(m, hops_constraint, 4A + 4B <= 160)
@constraint(m, malt_constraint, 35A + 20B <= 1190)

@constraint(m, profit_constraint, profit == 13A + 23B)

@objective(m, Max, profit)

solve(m)
```

```julia
println("Optimal Profit: \$", getvalue(profit))
println("Ale Produced: ", round(getvalue(A)) )
println("Beer Produced: ", getvalue(B))
```
